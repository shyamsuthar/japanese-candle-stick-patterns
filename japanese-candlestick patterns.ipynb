## Install packages

```python
import talib
import yfinance as yf
import pandas as pd
 
```

```python

data = yf.download("AAPL", start="2020-01-01", end="2022-03-15")
```

```python
data
```

```python
morning_star = talib.CDLMORNINGSTAR(data['Open'], data['High'], data['Low'], data['Close'])
```

```python
crows_2 = talib.CDL2CROWS(data['Open'], data['High'], data['Low'], data['Close'])
```

```python
data['Morning star'] = morning_star
data['2 crows'] = crows_2
```

```python
data
```

```python
morning_star_days = data[data['Morning star']!= 0]
```

```python
print(morning_star_days)
```

```python
doji = talib.CDLDOJI(data['Open'], data['High'], data['Low'], data['Close'])
data['Doji'] = doji
```

```python
doji_days = data[data['Doji']!= 0]
```
